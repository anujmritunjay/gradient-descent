```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 2 * np.random.rand(100, 1)  # square footage
y = 4 + 3 * x + np.random.randn(100, 1)  # price = 4 + 3 * sqft + noise
```

```python
m = 0
b = 0
learning_rate = 0.01
epochs = 5000
n = len(x)
```

```python
for epoch in range(epochs):
    y_pred = m * x + b
    error = y_pred - y
    
    # Calculating gradient
    m_grad = (2/n) * np.sum(error * x)
    b_grad = (2/n) * np.sum(error)

    # Updating parameters

    m -= learning_rate * m_grad
    b -= learning_rate * b_grad

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: m = {m}, b = {b}, MSE = {np.mean(error ** 2)}")
```

```python
plt.scatter(x, y, color='blue', label='Data Inputs')
plt.plot(x, m * x + b , color='red', label='Regression Line')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
def predict(val):
    return m * val + b
```
